# TRATAMENTO DA BASE FUNCIONARIOS

#Correção do valor numérico armazenado como texto com aspas e separador de milhar

```python
## Removendo ' " ', ',' , '.' e o alterando o tipo de dado
df_funcionarios['salario'] = (df_funcionarios['salario'].astype(str).str.replace('"', '', regex=False).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
)
df_funcionarios['salario'] = pd.to_numeric(df_funcionarios['salario'],errors='coerce')
display(df_funcionarios)
```

#Corrigindo valores incorretos na coluna 'salario':valores negativos e outlier extremo

```python
# Correção conforme Comunicado COM-2025-047:
# id_funcionario 11 (Gabriela Lima - Camareira): -2800,00 -> 1950,00
# id_funcionario 26 (Gabriela Lima - Manutenção): 999999,00 -> 2300,00

correcao = {
    11: 1950.00,
    26: 2300.00
}

for id_funcionario, salario in correcao.items():
    df_funcionarios.loc[
        df_funcionarios['id_funcionario'] == id_funcionario,
        'salario'
    ] = salario

display(
    df_funcionarios.loc[
        df_funcionarios['id_funcionario'].isin([11, 26]),
        ['id_funcionario', 'nome', 'salario']
    ]
)
```

#Padronização da coluna 'cargo' e 'departamento'

```python
df_funcionarios['cargo'] = (df_funcionarios['cargo'].str.strip().str.title())

df_funcionarios['departamento'] = (df_funcionarios['departamento'].str.strip().str.title())

mapa = {
    'A&B': 'Alimentos e Bebidas',
    'Alimentos E Bebidas' : 'Alimentos e Bebidas',
    'Admin.': 'Administração',
    'Gov.': 'Governança'
}
df_funcionarios['departamento'] = df_funcionarios['departamento'].replace(mapa)

display(df_funcionarios['cargo'].value_counts())
display(df_funcionarios['departamento'].value_counts())
```
